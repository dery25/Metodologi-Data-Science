#**Nomor 3**

```python
datapenjualan_makananpenutup ={
'Donat': 14,
'Pastel': 5,
'Kue Coklat': 12,
'Ice Cream': 19,
'Puding Vanila': 6,
'Brownies': 8,
'Puding Strawberi': 12,
'Puding Coklat': 9,
'Ice Cream Nutela': 10,
'Kue Coklat-Keju': 17,
'Kue Wajik': 2,
'Kue Sus': 9,
'Mochi': 13,
}


df = pd.DataFrame({
      'Jenis':datapenjualan_makananpenutup.keys(),
      'Jumlah':datapenjualan_makananpenutup.values()
})

df.sort_values(by='Jumlah', inplace=True)
x_coords = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
for i in range(3) :
  warna[i] = '#FF0000'


plt.figure(figsize=(10,5))
plt.title ('Popularitas Makanan Penutup')
plt.bar(
    x_coords, df['Jumlah'], tick_label=df['Jenis'], color=warna
)
plt.xticks(rotation = 90)
plt.ylabel('Jumlah Terjual Dalam 1 Minggu')
plt.xlabel('Jenis Makanan')
plt.show()
```

Sebutkan tiga makanan penutup yang harus
disingkirkan.

Berdasarkan data dan visualisasi di atas maka makanan penutup yang harus disingkirkan adalah :
1. Kue Wajik
2. Pastel
3. Puding Vanila
